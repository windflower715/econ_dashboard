```python
import pandas as pd 
import matplotlib.pyplot as plt 
from bbg import bdh, bdh_bulk
import matplotlib.pyplot as plt
import plotly
import plotly.express as px
import cufflinks 
pd.options.plotting.backend = "plotly"
import cufflinks as cf
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
init_notebook_mode(connected=True)

init_notebook_mode(connected=True)
cf.go_offline()
```

```python
df = bdh_bulk({'CAI':'GSUSCAI Index','FCI':'GSUSFCI Index',
               'SPX':'SPX Index','IG Index':'LBUSTRUU Index','10y yield':'USGG10YR Index','DXY':'DXY Curncy',
               'payroll':'NFP TCH Index','jobless claims':'INJCJC Index','unemployment rate':'USURTOT Index', 'wages yoy':'AHE MOM% Index',
               'bbg consumer comfort':'COMFCOMF Index','umichi consumer sentiment':'CONSSENT Index',
               'retail sales mom':'RSTAMOM  Index','retail sales yoy':'RSTAYOY Index','personal spending mom':'PCE CRCH Index',
               'cpi yoy index':'CPI YOY Index','cpi mom index':'CPI CHNG Index','inflation expectation':'T5YIFR Index',
               'manufacturing pmi':'MPMIUSMA Index','services pmi':'MPMIUSSA Index'},'20000101','20300101')
```

```python
def show_plot(title_input ,tkr_list,shape_input,start_date ):
    figdic = {}
    for i in tkr_list: 
        figdic[i] = df[[i]].dropna().loc[start_date:].figure(kind= 'line',title  =  title_input)

    cf.Figure(cf.subplots([figdic[x] for x in tkr_list],base_layout=figdic[tkr_list[0]].to_dict()['layout'],shape= shape_input)).iplot( )
```

# US Eon tracker

```python
title_input = 'Current activity index and financial condition index'
tkr_list = ['CAI','FCI']
show_plot(title_input,tkr_list, (2,1),'2005-01-01' )
```

```python
title_input = 'financial market'
tkr_list = ['SPX','IG Index','10y yield','DXY']
show_plot(title_input,tkr_list, (2,2),'2020-01-01' )
```

```python
title_input = 'Labor market'
tkr_list = [ 'payroll','jobless claims', 'unemployment rate', 'wages yoy']
show_plot(title_input,tkr_list, (2,2),'2020-01-01' )
```

```python
title_input = 'CPI'
tkr_list = [  'cpi yoy index',  'cpi mom index','inflation expectation']
show_plot(title_input,tkr_list, (1,3),'2020-01-01' )
```

```python
title_input = 'Consumption and consumer sentiment'
tkr_list = [ 'retail sales yoy', 'personal spending mom','bbg consumer comfort', 'umichi consumer sentiment']
show_plot(title_input,tkr_list, (2,2),'2020-01-01' )
```

```python
title_input = 'Business confidence'
tkr_list = [  'manufacturing pmi',
       'services pmi']
show_plot(title_input,tkr_list, (1,2),'2020-01-01' )
```
